```python
import pandas as pd
import numpy as np
path = r"C:\Users\kpras\Desktop\Book1.xlsx"
df = pd.read_excel(path,sheet_name=1)
print(df.head())
```

```python
df["Final_Trigger"] = (
    df["Final_Trigger"]
    .str.replace(r"\{.*?\}", "", regex=True)   # remove {anything}
    .str.replace(r"\s{2,}", " ", regex=True)   # collapse extra spaces
    .str.strip()
)

agg = (
    df
    .groupby(["Final_Severity", "RuleID", "Final_Trigger"])
    .size()
    .reset_index(name="Count of Trigger")
)

top10 = (
    agg
    .sort_values(
        ["Final_Severity", "Count of Trigger"],
        ascending=[True, False]
    )
    .groupby("Final_Severity")
    .head(10)
)
```

```python
final_rows = []

for severity, group in top10.groupby("Final_Severity", sort=False):
    # Severity header row
    final_rows.append({
        "RuleID": "",
        "Final_Trigger": severity,
        "Count of Trigger": ""
    })
    
    # Actual trigger rows
    for _, row in group.iterrows():
        final_rows.append({
            "RuleID": row["RuleID"],
            "Final_Trigger": row["Final_Trigger"],
            "Count of Trigger": row["Count of Trigger"]
        })

final_df = pd.DataFrame(final_rows)
```
